```python
from pathlib import Path
import pandas as pd
import torch
import re
```

```python
result = pd.DataFrame()

for file in Path().rglob('model_final.pt'):
    tmp = torch.load(file, map_location=torch.device('cpu'))['hist']
    tmp = pd.DataFrame.from_records(tmp)

    file = str(file)
    alpha = None
    if 'alpha' in file:
        alpha = re.findall('alpha-[0-9].{2,3}', file)[0]
        # alpha = float(alpha[-4:]) * 100
        alpha = float(alpha[-4:])

    sample_size = re.findall('train[0-9].{2,3}', file)[0]
    sample_size = sample_size[-4:]
    _, dataset, arch, *_ = re.findall('[a-z0-9]+', file)
    
    # print(file, end='\t')
    method = re.findall('(?:[a-z]_?)+',file)
    print(method)
    # method = method[9] if alpha in method else method[11]
    method = method[8] if alpha in method else method[11]
    
    if method == 'None':
        sample_size = 1
    # tmp['alpha'] = int(alpha)
    tmp['alpha'] = alpha
    tmp['sample_size'] = float(sample_size)
    tmp['dataset'] = dataset
    tmp['arch'] = arch
    tmp['method'] = method
    tmp['epoch'] = range(len(tmp))
    result = pd.concat([tmp, result])
# result[result.method.isna()] = 'full dataset'
result = result[result.epoch<20]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, len(result.alpha.unique()), figsize=(20,8))
# fig, axes = plt.subplots(1, 1)

for ax, a in zip(axes, result.alpha.unique()):
    print(a)
    if not a == "nan":
        sns.lineplot(
            result[ (result.alpha == a) | (result.alpha.isna()) ], 
            x='epoch', 
            y='train_acc', 
            hue='method',
            ax=ax
        )
        ax.set_title(a)

# plt.tight_layout()
plt.show()
```
