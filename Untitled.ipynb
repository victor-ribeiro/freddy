```python
from pathlib import Path
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import torch
import re
```

```python
plt.rcParams['figure.figsize'] = 22, 8
```

```python
result = pd.DataFrame()

for file in Path().rglob('model_final.pt'):
    f = torch.load(file, map_location=torch.device('cpu'))
    tmp = f['hist']
    tmp = pd.DataFrame.from_records(tmp)
    file = str(file)
    alpha = None
    if 'alpha' in file:
        alpha = re.findall('alpha-[0-9].{2,3}', file)[0]
        alpha = float(alpha[-4:])

    sample_size = re.findall('train[0-9].{2,3}', file)[0]
    sample_size = sample_size[-4:]
    _, dataset, arch, *_ = re.findall('[a-z0-9]+', file)
    
    # print(file, end='\t')
    method = re.findall('(?:[a-z]_?)+',file)
    # print(method)
    # method = method[9] if alpha in method else method[11]
    print(method[9], alpha, sample_size)
    method = method[11] if 'alpha' in method else method[9]
    
    if method == 'None':
        sample_size = 1
    # print(f['grad_norm'])
    tmp['epoch'] = range(len(tmp))
    tmp['alpha'] = float(alpha) if alpha else None
    tmp['dataset'] = dataset
    tmp['arch'] = arch
    tmp['sample_size'] = float(sample_size)
    tmp['method'] = method
    result = pd.concat([tmp, result])
# result[result.method.isna()] = 'full dataset'
result = result[result.epoch<20]
```

```python


# fig, axes = plt.subplots(1, 1)

for i in result.loc[result.alpha.notna(), 'alpha'].unique():
    print(i)
    sns.lineplot(
        result [
            (result.alpha == i) |
            (result.alpha.isna()) 
        ], 
        x='epoch', 
        y='val_acc', 
        hue='method',
        style='method',
        alpha=.7
        # ax=ax
    )
    plt.ylim([0, .75])


# plt.tight_layout()
    plt.show()
```
