```python
from pathlib import Path
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import torch
import re
```

```python
plt.rcParams['figure.figsize'] = 22, 8
```

```python
result = pd.DataFrame()
# 
for file in Path('outputs').rglob('model_final.pt'):
# for file in Path('outputs').rglob('model_epoch_399'):
# for file in Path().rglob('model_epoch_5799.pt'):
    file = str(file)
    alpha = None
    if 'alpha' in file:
        alpha = re.findall('alpha-[0-9].{2,3}', file)[0]
        alpha = float(alpha[-4:])

    sample_size = re.findall('train[0-9].{2,3}', file)[0]
    sample_size = sample_size[-4:]
    _, _, _, dataset, arch, *_ = re.findall('[a-z0-9]+', file)
    method = re.findall('(?:[a-z]_?)+',file)
    method = method[11] if 'alpha' in method else method[9]
    print(method)
    # if not (method == 'freddy' or method == 'random'):
    #     continue
    print( alpha, sample_size)
    
    f = torch.load(file, map_location=torch.device('cpu'))
    try:
        plt.hist(f['selected'], bins=50)
        plt.yscale('log')
        # plt.axhline(5000)
        plt.show()
        #############
        selected = f['epoch_selection']
        #############

        # plt.scatter(f['selected'], f['importance'])
        plt.scatter(f['reaL_error'], f['importance'])
        plt.show()
    except:
        pass
    tmp = f['hist']
    tmp = pd.DataFrame.from_records(tmp)
    if method == 'None':
        sample_size = 1
    # print(f['grad_norm'])
    # tmp['epoch'] = range(len(tmp))
    tmp['alpha'] = float(alpha) if alpha else None
    tmp['dataset'] = dataset
    tmp['arch'] = arch
    tmp['sample_size'] = float(sample_size)
    tmp['method'] = method
    result = pd.concat([tmp, result])
# result[result.method.isna()] = 'full dataset'
# result = result[result.epoch<20]
result
```
